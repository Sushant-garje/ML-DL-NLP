```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_diabetes
```

```python
dia = load_diabetes()
```

```python
dia
```

```python
X = pd.DataFrame(dia["data"],columns=dia['feature_names'])
```

```python
X.head()
```

```python
y = dia["target"]
```

```python
y
```

```python
X.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(X.corr(), annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
X_train,y_train
```

```python
from sklearn.tree import DecisionTreeRegressor
dtreeReg = DecisionTreeRegressor()
dtreeReg.fit(X_train,y_train)
```

```python
regressor = DecisionTreeRegressor()
```

```python
param={
    'criterion':['squared_error','friedman_mse','absolute_error'],
    'splitter':['best','random'],
    'max_depth':[1,2,3,4,5,10,15,20,25],
    'max_features':['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(estimator=regressor,param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,root_mean_squared_error,r2_score,mean_absolute_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
model = DecisionTreeRegressor(criterion="squared_error",max_depth=4,max_features="log2",splitter="best")
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
print(root_mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
```

```python
##Visualize the Decision Tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(model,filled=True)
plt.show()
```
